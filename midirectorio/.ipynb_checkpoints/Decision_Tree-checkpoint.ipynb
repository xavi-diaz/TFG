# Árbol de decisiones
como ejemplo del aprendizaje en máquina, se usa un subconjunto de la base de datos.

*importaciones*

```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error #para calcular la eficacia del árbol.
from sklearn.tree import export_graphviz #para ver el árbol
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

*obtención y tratamiento de la base de datos*

```python
df = pd.read_csv('scaled_df.csv')


#se usa un subconjunto de la base de datos completa pues este es solo un ejemplo ilustrativo.

subconjunto = df.sample(frac=0.4, random_state=42)
```

**construcción de tabla profundidad/error**

```python
depths = [3, 5, 7, 10, 100, 300, 1000, 3000, 5000, 7000]
mae_values = []

results_df = pd.DataFrame(columns=["Profundidad", "Error Absoluto Medio"])

target = 'peakwavs_max'
features = [col for col in df.columns if col not in ['smiles', 'solvent', 'source', target]]

# Dividir en conjunto de entrenamiento y prueba
x = subconjunto[features]
y = subconjunto[target]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

for depth in depths:
    print(f"Entrenando con profundidad: {depth}")
    reg = DecisionTreeRegressor(max_depth=depth, criterion='absolute_error', max_leaf_nodes=None)

    reg.fit(x_train, y_train)
    y_pred = reg.predict(x_test)

    mae = mean_absolute_error(y_test, y_pred)
    mae_values.append(mae)

    # Agregar resultados al DataFrame
    results_df = pd.concat([results_df, pd.DataFrame({"Profundidad": [depth], "Error Absoluto Medio": [mae]})], ignore_index=True)

# Mostrar tabla en formato DataFrame
print(results_df)
results_df.to_csv('/Resultados/decision_tree/tabla_trees.csv')
```

```python

resultados = pd.read_csv('tabla_trees.csv')

depths = resultados['Profundidad']
mae_values = resultados['Error Absoluto Medio']
# Graficar Error Absoluto Medio vs Profundidad
plt.figure(figsize=(5, 5))
plt.plot(depths, mae_values, marker='o', markersize=6, linestyle='-', color='k', label="Error Absoluto Medio")
plt.xlabel("Profundidad del Árbol")
plt.ylabel("Error Absoluto Medio (MAE)")
plt.title("Error Absoluto Medio vs Profundidad del Árbol")
plt.xscale("log")
plt.grid(True, which="major", linestyle="solid", linewidth=0.5)
plt.savefig('Rend_Decision_trees.png', bbox_inches='tight')
plt.show()
```

**constucción de un arbol de profundidad establecida**

```python
depth = int(input("Introduce la profundidad del árbol: "))

# Create Decision Tree classifer object
decision_tree = DecisionTreeRegressor(max_depth = depth,
                        criterion='absolute_error')

# selección de variables objetivos y separación de los datos
target = 'peakwavs_max'
features = [col for col in df.columns if col not in ['smiles', 'solvent','source', target]]

# Dividir en conjunto de entrenamiento y prueba
x = df[features]
y = df[target]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# entrenamiento
decision_tree = decision_tree.fit(x_train, y_train)

#Predicción
y_pred = decision_tree.predict(x_test)

# Cálculo de la exactitud del modelo
mae = mean_absolute_error(y_test, y_pred)
print(f'El árbol con profundidad: {depth} tiene un error absoluto medio igual a: {mae:.4f}')

# Exportar el árbol en formato DOT
file_name = f'tree_{depth}.dot'
export_graphviz(decision_tree,
                out_file=file_name,
                feature_names=features,
                filled=True,
                rounded=True,
                special_characters=True)
```
